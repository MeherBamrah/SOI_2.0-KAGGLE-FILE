```python
import requests
from kaggle_secrets import UserSecretsClient

secrets = UserSecretsClient()
KEY = secrets.get_secret("LANGUAGE_KEY")
ENDPOINT = secrets.get_secret("LANGUAGE_ENDPOINT")

def analyze(kind, text):
    url = f"{ENDPOINT}language/:analyze-text?api-version=2023-04-01"
    headers = {"Ocp-Apim-Subscription-Key": KEY, "Content-Type": "application/json"}
    body = {
        "kind": kind,  # "SentimentAnalysis" | "KeyPhraseExtraction" | "EntityRecognition"
        "analysisInput": {"documents": [{"id": "1", "language": "en", "text": text}]}
    }
    return requests.post(url, headers=headers, json=body).json()

review = "THE FOOD IS DISSOPOINTING BUT THE INTERIOR OF THE RESTURANT WAS CLASSY AND CHICK."

print(analyze("SentimentAnalysis", review))
print(analyze("KeyPhraseExtraction", review))
print(analyze("EntityRecognition", review))
```
